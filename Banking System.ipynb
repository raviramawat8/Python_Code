```python
from getpass import getpass
import time
bank = { 
        'user' : [ 'ram', 'sam','vijay' ],
        'acc' : [ 1001, 1002, 1003 ],
        'password' : [ 'grras@123', 'something', 'redhat'],
        'balance' : [ 10000.23, 20000.45, 124322.234]
        }
admin = 'root'
admin_password = 'redhat@123'
import json
f = open("C:\\Training\\Project_1\\bank.db",'w+')
json.dump(bank,f)

while True :
    print("\n\n")
    print("*"*50)
    print("Welcome to Python Bank")
    print("*"*50)
    print("\n\n1. Login \n2. Signup \n3. Admin \n4. Exit")
    ch = int(input("Your Choice : "))
    
    if ch == 1 :
        print("\n\n")
        uacc = int(input("Acc Number : "))
        if uacc in bank['acc'] :
            upass = getpass("\nPassword : ")
            index = bank['acc'].index(uacc)
            f.write("\n\nLoging in : \nAccount no = {} \n Password = {} \n Time = {}".format(bank['acc'][index],bank['password'][index],time.ctime()))
            if upass == bank['password'][index]:
                while True :
                    print("\n\n1. Debit\n2. Credit\n3. Check Balance\n4. Change Password \n5. Logout")
                    ch = int(input("Choice : "))
                    if ch == 1 :
                        print("\n\n")
                        deb = float(input("Enter the Amount you want to debit\n"))
                        if deb > bank['balance'][index]:
                            print("\n\n You have insufficient balance")
                        else:
                            bank['balance'][index] -= deb
                            print("\n\nThe amount has been successfully debited to your Account\n")
                            print("\nNow Your Main Balance is {}".format(bank['balance'][index]))
                            f.write("\n\nAccount Balance Debited = {}\n Time = {}".format(bank['balance'][index],time.ctime()))
                    elif ch == 2 :
                        print("\n\n")
                        cred = float(input("Enter the Amount you want to credit"))
                        bank['balance'][index] += cred
                        print("\n\nThe amount has been successfully credited to your Account\n")
                        print("\nNow Your Main Balance is {}".format(bank['balance'][index]))
                        f.write("\n\nAccount Balance Credited = {}\n Time = {}".format(bank['balance'][index],time.ctime()))
                    elif ch == 3 :
                        print("\n\nYour Balance for account no.{} is {}".format(bank['acc'][index],bank['balance'][index]))
                        f.write("\n\nAccount Balance = {}".format(bank['balance'][index]))
                    elif ch == 4 :
                        print("\n\n")
                        oldpass = getpass("\n\nEnter your old Password\n")
                        if oldpass == bank['password'][index]:
                            newpass = getpass("\n\nEnter your new Password")
                            bank['password'][index]=newpass
                            print("\n\nYour Password has been successfully changed")
                            f.write("\n\nPassword Changed \n New Password = {}\n Time = {}".format(bank['password'][index],time.ctime()))
                        else:
                                print("\n\n Wrong Password\n\nPlease Try Again")
                    elif ch == 5 :
                        print("\nLogging you out.....\n")
                        f.write("\nLoging Out")
                        break
                    else :
                        print("\n\nInvalid Choice ")
                        print("Try Again")
            else :
                print("\n\nInvalid Password\n\nTry Again")
            
        else :
            print("\n\nNo such Account Exists \nYou should signup\n\n")
    elif ch == 2 :
        acc = bank['acc'][-1] + 1
        name = input("Username : ")
        bal = float(input("Initial Blance : "))
        password = getpass("Set Password : ")
        bank['user'].append(name)
        bank['acc'].append(acc)
        bank['password'].append(password)
        bank['balance'].append(bal)
        print("\nAccount Sucessfully Created")
        print("Note down your account num :  {}".format(acc))
        print("Now you can login\n")
        f.write("\n\nSigning up:\n Initial Balance = {} \n User_Name = {} \n Account no. = {} \n Password ={} \n Time = {}".format(bank['balance'][-1], bank['user'][-1],bank['acc'][-1], bank['password'][-1],time.ctime()))
    elif ch == 3 :
        username = input("\nEnter name : ").strip().lower()
        password = getpass()
        if username == admin : 
            if password == admin_password :
                print()
                for key,value in bank.items():
                    print("{}={}".format(key,value))
                f.write("\nAdmin Login : \n User_Name = {} \n Password = {}\n Time = {}".format(username,password,time.ctime()))
            else :
                print("\nInvalid password \nTry Again")
        else :
            print("\nThis user is not an admin.")
    elif ch == 4 :
        ch = input("\n\nDo really want to exit ? (y/n) : ").strip().lower()
        if ch == 'y' or ch == 'yes' :      
            print("\nThanks for using our services ")
            f.write("Exited")
            break
    else :
        print("\nInvalidChoice\nTry again")
f.close()
```
